Responda las siguientes preguntas:

> d) De todas las combinaciones que entrenó, ¿cual es la que mejores resultados le da? Para esta pregunta evalue el modelo con el conjunto de validación de la grilla utilizando la métrica de evaluación $mean\_squared\_error$.

**Nota:** Recuerde que estos experimentos debe realizarlos para ambas representaciones de los conjuntos ($A$ y $B$).



````python
def aux_report(X, y, model, scaler):
    y_pred = scaler.inverse_transform(np.array(model.predict(X)).reshape(-1, 1))
    best_params = model.best_params_
    error = mean_squared_error(y, y_pred)
    print(f"Mean squared error: {error}")
    print(f"Params:")
    for key in best_params.keys():
        print(f"    {key} = {best_params[key]}")
    print("")
    return error
    
def report(X, y, model_a, model_b, scaler):
    print("A (Full)")
    error_a = aux_report(X, y, model_a, scaler)
    print("B (Reduced)")
    error_b = aux_report(X[selected_columns], y, model_b, scaler)
    if error_a < error_b:
        print("best representation: A (Full)")
    elif error_b < error_a:
        print("best representation: B (Reduced)")
    else:
        print("best representation: Both")
````

````python
param_grid = {'max_depth': [2, 20, 200], 
              'max_leaf_nodes': [10, 50, 100]}

tree_regressor = DecisionTreeRegressor()

tree_regressor_GS_a = GridSearchCV(
            tree_regressor, 
            param_grid, 
            scoring=mean_squared_error,               
            refit=True, 
            n_jobs = 8)

tree_regressor_GS_b = GridSearchCV(
            tree_regressor, 
            param_grid, 
            scoring=mean_squared_error,               
            refit=True, 
            n_jobs = 8)
````

````python
model_1_a = tree_regressor_GS_a.fit(X_train_a,y_train_a)
model_1_b = tree_regressor_GS_b.fit(X_train_b,y_train_b)
````

````python
report(X_test, y_test, model_1_a, model_1_b, y_scaler)
````

> e) Entrene el conjunto de entrenamiento completo con el mejor modelo y mejor representación ($A$ o $B$) obtenida y evalúe su desempeño sobre el conjunto de prueba utilizando la métrica de evaluación $mean\_squared\_error$.

````python
best_tree_regressor = model_1_a
y_pred = y_scaler.inverse_transform(np.array(best_tree_regressor.predict(X_test)).reshape(-1, 1))
print(mean_squared_error(y_pred, y_test))
````

### 1.2.2 Random Forest Regressor

Entrene un bosque aleatorio de regresión manteniendo fijos por defecto los que no se mencionen y variando los siguientes parámetros utilizando el módulo gridsearchcv de la librería sklearn:

* max_depth: Probar con las profundidades 2, 20, 200. Describa brevemente qué es lo que significa modificar este hiperparámetro.

* max_samples: Probar con las siguientes proporciones 0.1, 0.5, 1.0. Describa brevemente qué es lo que significa modificar este hiperparámetro.

El objetivo de los métodos $\textit{ensemble machine}$ es combinar las predicciones de varios estimadores base construidos con un algoritmo de aprendizaje determinado para mejorar la capacidad de generalización / robustez respecto a un único estimador. Se suelen distinguir dos familias de métodos ensemble: los métodos $averaging$ y los métodos $boosting$. Respecto a lo anterior responda las siguientes preguntas:

```max_depth``` Este hiperparámetro controla la profundidad máxima que los árboles en el Random Forest pueden alcanzar. Si elegimos un valor pequeño como 2, los árboles solo podrán hacer dos divisiones, lo que puede llevar a un modelo demasiado simple y a un bajo rendimiento debido al underfitting. Si elegimos un valor muy grande como 200, los árboles podrán hacer muchas divisiones y pueden capturar demasiado bien las peculiaridades y el ruido en los datos de entrenamiento, lo que puede llevar a un sobreajuste y a un mal rendimiento en los datos no vistos.

```max_samples``` Este hiperparámetro determina la fracción de muestras a utilizar en cada árbol. Si se establece en 0.1, solo se utilizará el 10% de las muestras para entrenar cada árbol. Si se establece en 1.0, se utilizarán todas las muestras. El uso de una fracción más pequeña puede hacer que el modelo sea más robusto a las variaciones en los datos y menos propenso a sobreajustarse, pero también podría hacer que el modelo sea menos preciso si la fracción es demasiado pequeña.
> a) Defina con sus propias palabras qué son y en qué se diferencian los métodos $averaging$ y $boosting$.
Los métodos de conjunto son técnicas que combinan múltiples modelos de aprendizaje automático para tratar de producir un modelo más poderoso y preciso.

**Métodos de Averaging** trabaja entrenando varios modelos independientes y luego combinando sus predicciones promediándolas bajo un metodo. El objetivo es reducir la varianza del modelo, especialmente si los modelos individuales tienen una alta varianza. Un ejemplo de este enfoque es el algoritmo de Random Forest.

**Métodos de Boosting** entrenan los modelos de manera secuencial, donde cada modelo nuevo trata de corregir los errores realizados por el conjunto de modelos ya entrenados. El objetivo aquí es reducir el sesgo del modelo y crear un modelo que sea capaz de aprender de los errores y mejorar con cada iteración. Un ejemplo de este enfoque es el algoritmo AdaBoost.
> b) A cual de los dos anteriormente definidos pertenece el modelo de regresión Random Forest?

El modelo de Random Forest pertenece a los métodos de Averaging. Cada árbol en el bosque se entrena de manera independiente y sus predicciones son luego promediadas para obtener la predicción final del bosque.
> c) De todas las combinaciones que entrenó, ¿cual es la que mejores resultados le da? Para esta pregunta evalue el modelo con el conjunto de validación utilizando la métrica de evaluación $mean\_squared\_error$.

````python
model_2_a = forest_regressor_GS_a.fit(X_train_a,y_train_a)
model_2_b = forest_regressor_GS_b.fit(X_train_b,y_train_b)
````

* max_samples: Probar con las siguientes proporciones 0.1, 0.5, 1.0. Describa brevemente qué es lo que significa modificar este hiperparámetro.

```n_estimators``` Este hiperparámetro se refiere al número de árboles a utilizar en el algoritmo. Cada uno de estos árboles será un estimador independiente obtenido a partir de una muestra de los datos de entrenamiento. Incrementar el número de estimadores permitirá al algoritmo tener una "visión más completa" del espacio de datos de entrenamiento, pero esto también incrementará el tiempo de cálculo y puede llegar a un punto de rendimientos decrecientes donde añadir más estimadores no mejora significativamente el rendimiento del modelo.

```max_samples``` Este hiperparámetro indica la proporción de muestras a utilizar para entrenar cada estimador individual. Al igual que en el Bosque Aleatorio, si se selecciona una fracción menor de las muestras para entrenar a cada estimador, se está introduciendo más diversidad entre los estimadores, lo que puede hacer que el modelo final sea más robusto.
Responda las siguientes preguntas

> a) ¿Cómo funciona el modelo Bagging?

El modelo de Bagging funciona construyendo varios estimadores independientes sobre subconjuntos aleatorios del conjunto de datos de entrenamiento haciendo muestreo Bootstrap y luego agregando sus predicciones. Este agregado puede ser un promedio en el caso de la regresión o una votación en el caso de la clasificación. Al combinar predicciones de varios estimadores, se puede reducir la varianza y mejorar la robustez del modelo.
> b) A qué familia de métodos pertenece el Bagging regresor: $averaging$ o $boosting$.

El Bagging Regressor pertenece a los métodos de Averaging. El cual funciona construyendo varios estimadores de manera independiente y luego promediando sus predicciones. Cada estimador se entrena con una muestra distinta del conjunto de datos, lo que ayuda a mejorar la robustez del modelo final.
> c) De todas las combinaciones que entrenó, ¿cual es la que mejores resultados le da? Para esta pregunta evalue el modelo con el conjunto de validación utilizando la métrica de evaluación $mean\_squared\_error$.

## 2.2 Modelos de clasificación

### 2.2.1 Redes Neuronales

> a) ¿Cómo se entrena una red neuronal feedforward?

Una red neuronal feedforward se entrena mediante un proceso iterativo que sigue tres pasos principales: forward pass, cálculo de la pérdida y backpropagation.
1. **Forward pass:** en este paso, la red neuronal toma los datos de entrada y los pasa a través de todas sus capas, desde la capa de entrada hasta la capa de salida. Cada neurona en cada capa aplica una transformación a los datos (mediante una función de activación y los pesos de importancia) y pasa el resultado a las neuronas de la siguiente capa.
2. **Cálculo de la pérdida:** una vez que los datos han pasado a través de toda la red, se genera una predicción en la capa de salida. Esta predicción se compara con el valor objetivo utilizando una función de pérdida, que calcula qué tan lejos está la predicción del valor objetivo.
3. **Backpropagation:** por ultimo, la red neuronal usa el algoritmo de backpropagation para calcular el gradiente de la función de pérdida con respecto a cada uno de los pesos de la red. Luego, estos gradientes se utilizan para actualizar los pesos de la red en una dirección que minimice la pérdida.
Este proceso se repite múltiples veces (en cada epoch de entrenamiento), ajustando los pesos de la red a través de los gradientes y el optimizador hasta que la red esté suficientemente entrenada.
> b) ¿Qué es una neurona en este contexto?

En el contexto de las redes neuronales, una neurona es una unidad de cálculo individual, es de alguna forma una función que toma una o más entradas, realiza alguna transformación en ellas y produce una salida. Las entradas son ponderadas por los pesos y se suma un sesgo. La suma ponderada se pasa a través de una función de activación que determina la salida de la neurona. En el modelo de una red neuronal, las neuronas se organizan en capas y las salidas de una capa se convierten en las entradas de la siguiente y así sucesivamente.
> c) ¿Qué es una función de activación y para que sirven?

Una función de activación es una función matemática que se aplica a la salida de una neurona en una red neuronal. Estas funciones tienen una enorme importancia en las redes neuronales ya que introducen no linealidades en el modelo ampliando enormemente su capacidad de adaptación, lo que permite a la red neuronal aprender relaciones complejas entre las entradas y las salidas. Sin estas funciones de activación, las redes neuronales se reducirían a modelos lineales, independientemente de cuántas capas se apilen en la red ya que al final terminaria colapsando la red en si misma debido a los teoremas matematicos que dictan que muchos modelos lineales equivalen a uno solo. Algunas funciones de activación comunes incluyen la función sigmoide, tangente hiperbólica y ReLU.
> d) ¿A qué se le llama capa densa?

Una capa densa, también conocida como capa totalmente conectada, es una capa en una red neuronal donde cada neurona está conectada a todas las neuronas en la capa anterior y a todas las neuronas en la capa siguiente. En otras palabras, las neuronas en una capa densa tienen "conexiones completas" a todas las activaciones en la capa anterior, por lo que se realizan operaciones de producto punto entre las entradas y los pesos, sumado el sesgo.
> e) Si usted conoce la dimensión entrada, la de salida, la cantidad de capas y neuronas de cada capa, ¿cómo se calcula la cantidad de pesos de la red?

La cantidad total de pesos en una red neuronal feedforward se puede calcular sumando los pesos de todas las capas. Para una capa individual, el número de pesos es igual al número de neuronas en la capa actual multiplicado por el número de neuronas en la capa anterior. También se tiene en cuenta que cada neurona tiene un sesgo, así que este debe ser sumado a la par. Así es como para una capa con n neuronas que se conecta a una capa anterior con m neuronas, habrá m x n pesos más n sesgos, para un total de m x n + n parámetros entrenables.
> f) ¿A qué se refiere el fenómeno de gradiente desvaneciente?, ¿y el gradiente explosivo?

El problema del gradiente desvaneciente y el gradiente explosivo son dos problemas comunes que se suelen presentar durante el entrenamiento de las redes neuronales profundas.

El gradiente desvaneciente ocurre cuando los gradientes se vuelven muy pequeños a medida que se retropropagan hacia las capas inferiores de la red. Esto significa que los pesos en las primeras capas de la red cambian muy poco durante el entrenamiento, lo que puede llevar a que la red deje de aprender y el entrenamiento se estanque.

Por otro lado, el problema del gradiente explosivo ocurre cuando los gradientes se vuelven muy grandes, lo que puede llevar a actualizaciones de peso demasiado grandes y a una red que es inestable y no puede aprender de los datos de entrada.

Estos problemas se pueden solucionar aplicando algunas funciones diseñadas para reducir los efectos tales como la funcion de activación ReLU, inicializaciones de peso especializadas y regularización.
> g) Para esta tarea, como debería ser la capa de salida de la red considerando que previemente aplicó una transformación One Hot Encoding a las clases del conjunto?

````python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.activations import relu, tanh, sigmoid, softmax
````

````python
def shallow_ANN(k, activation):
    input_layer = Input(shape = (x_train.shape[1:]))
    hidden_layer = Dense(units=k, activation=activation)(input_layer)
    output_layer = Dense(units=6, activation=softmax)(hidden_layer)
    return Model(inputs=input_layer , outputs=output_layer)
````

````python
ANN_tanh = shallow_ANN(32, tanh)
ANN_sigmoid = shallow_ANN(32, sigmoid)
ANN_relu = shallow_ANN(32, relu)
````

* ¿Qué significa el batch_size?

```batch_size``` se refiere al número de muestras de entrenamiento utilizadas en una iteración de actualización del modelo durante el entrenamiento de redes neuronales y otros algoritmos de aprendizaje automático.

Se puede decir que es el número de ejemplos de entrenamiento que se utilizan en una sola iteración de aprendizaje. Durante cada iteración, el modelo calcula el gradiente de la función de pérdida con respecto a los pesos del modelo basándose en el conjunto actual de ejemplos y luego actualiza los pesos.

El tamaño del lote puede tener un impacto importante en la eficiencia y la calidad del entrenamiento del modelo. para ello es necesario tomar algunas consideraciones tales como mantener un tamaño de **lote más grande** que puede permitir al modelo aprovechar las optimizaciones de hardware y computar el gradiente más rápidamente, pero también puede requerir más memoria y así convergiendo más rapidamente pero con soluciones subóptimas. Un tamaño de **lote más pequeño** puede resultar en actualizaciones más ruidosas del modelo y a una convergencia más lenta, lo que en algunos casos puede ayudar al modelo a escapar de mínimos locales de la función de pérdida lo que se traduce en mejores soluciones.

### 2.2.2 AdaBoost y arbustos

> a) Describa brevemente como funciona el modelo AdaBoost.

AdaBoost o Adaptive Boosting es un algoritmo de aprendizaje supervisado que se utiliza para mejorar el rendimiento de los modelos de aprendizaje automático. El objetivo de AdaBoost es combinar las predicciones de varios modelos base que suelen ser denominados "estimadores débiles" para crear un modelo de conjunto robusto y preciso para alcanzar a ser un "estimador fuerte".

Este algoritmo funciona por varias rondas. En cada ronda, se añade un nuevo estimador débil al modelo de conjunto. Este estimador se entrena para corregir los errores cometidos por el modelo de conjunto actual. Específicamente, aumenta el peso de las muestras que fueron clasificadas incorrectamente en la ronda anterior, lo que obliga al nuevo estimador a centrarse en estas muestras difíciles, luego se combinan las predicciones de todos los estimadores débiles en una predicción final. Cada estimador débil tiene un peso que determina cuánto contribuye a la predicción final. Los estimadores que tienen un buen rendimiento en muestras difíciles reciben un peso mayor.
> b) ¿Qué clasificador se puede usar junto con Adaboost?

Puede utilizarse muchos modelos base o clasificadores débiles. Pero se suele utilizar árboles de decisión de un solo nivel debido a su baja complejidad y eficiencia. Aunque estos árboles son modelos muy simples y de alto sesgo, AdaBoost puede combinar eficazmente muchos de ellos para crear un modelo de conjunto de baja varianza y bajo sesgo. Otros tipos de modelos base también pueden usarse, siempre que puedan manejar la reponderación de las muestras.

> c) ¿Cómo funciona el _learning rate_ en el modelo AdaBoost?

El learning rate (que significa la tasa de aprendizaje) en AdaBoost es un parámetro que controla cuánto se ajustan los pesos de las muestras después de cada ronda de boosting. Un learning rate más alto significa que los pesos de las muestras se ajustarán más después de cada ronda, haciendo que el algoritmo se centre más en las muestras que fueron clasificadas incorrectamente. Un learning rate más bajo suaviza este efecto, lo que puede hacer que el modelo sea más robusto a los outliers pero también puede ralentizar la convergencia del algoritmo.

Además, el learning rate también afecta a los pesos de los estimadores débiles en el modelo de conjunto. Un learning rate más alto aumentará la diferencia de peso entre los estimadores que se desempeñan bien en muestras difíciles y los que no lo hacen.
